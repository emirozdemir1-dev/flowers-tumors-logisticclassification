Libraries

```python
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
warnings.filterwarnings('ignore')

import pandas as pd
from sklearn.linear_model import LogisticRegression
```

Load the data in a csv file into a dataframe

```python
# the data set is available at the url below.
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-BD0231EN-SkillsNetwork/datasets/iris.csv"

# using the read_csv function in the pandas library, we load the data into a dataframe.

df = pd.read_csv(URL)

# show 5 random rows from the dataset
df.sample(5)

df.shape

df.Species.value_counts().plot.bar()
```

Identify the target column and the feature columns

```python
target = df["Species"]  

features = df[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
```

Build and train a classifier

```python
classifier = LogisticRegression() # Create an instance of the LogisticRegression class

classifier.fit(features,target) # Fit the model to the data
```

Evaluate the model

```python
# Higher the score, better the model.
classifier.score(features,target) # 0.9733333333333334

classifier.predict([[5.4,2.6,4.1,1.3]]) # Predicting the species of a flower with 5.4 cm sepal length, 2.6 cm sepal width, 4.1 cm petal length, and 1.3 cm petal width
# # it should be a Iris-versicolor 
```

Exercise Chapter 1: Load Dataset

```python
URL2 = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-BD0231EN-SkillsNetwork/datasets/cancer.csv"

df2 = pd.read_csv(URL2)

df2.head()
```

Exercise Chapter 2: Identify the target column and the data columns

```python
target = df2["diagnosis"]  # Identify the target column
features = df2.drop(columns=["diagnosis"])  # Identify the feature columns by dropping the target column
```

Exercise Chapter 3: Build and Train a new classifier

```python
classifier2 = LogisticRegression()  # Create an instance of the LogisticRegression class for the cancer dataset
classifier2.fit(features, target)  # Fit the model to the cancer dataset
```

Exercise Chapter 4: Evaluate the model

```python
classifier2.score(features, target)  # Evaluate the model's performance on the cancer dataset
```
